```python
from lib import ENIGH_Data
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import os
```

```python
sns.set_style("whitegrid", {"grid.linestyle":"--","axes.edgecolor":"0"})
```

## Figure 2-4 thesis

Figure that describe the number of variables according to their proportion of null values

```python
yr = 2020 #Options 2016,2018,2020

ENIGH = ENIGH_Data(year=yr, clean=False)
ENIGH_dataset = ENIGH.read_data()

f = plt.figure(figsize=(10,6))
f.subplots_adjust(hspace=0.2)
fontsize_ticks = 20
gs = gridspec.GridSpec(1,1)
ax = plt.subplot(gs[0])
ax.tick_params(labelsize=fontsize_ticks)
sns.despine()

sns.histplot(ENIGH.proportion_nan(ENIGH_dataset), kde=False)

plt.xlabel("$n_{i}$", fontsize=20)
plt.ylabel("Frecuencia", fontsize=20)

plt.savefig(os.getcwd()+"/figures/NullValues"+str(yr)+".pdf", bbox_inches="tight")

plt.show()
```
